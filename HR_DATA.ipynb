# Data Analysis

```python
hr.isnull().sum()
```

```python
hr.Attrition = hr.Attrition.map({'Yes':1,'No':0})
```

```python
hr.BusinessTravel = hr.BusinessTravel.map({'Travel_Frequently':1, 'Travel_Rarely':2,'Non-Travel':3})
```

```python
hr.EducationField = hr.EducationField.map({'Life Sciences':5, 'Medical':4, 'Marketing':3, 'Technical Degree':2, 'Human Resources':1,'Other':0})
```

```python
hr.Department = hr.Department.map({'Research & Development':2, 'Sales':1, 'Human Resources':0})
```

```python
hr.Gender = pd.get_dummies(hr.Gender, drop_first=True)
```

```python
hr.JobRole = hr.JobRole.map({'Sales Executive':8,'Research Scientist':7,'Laboratory Technician':6,
                                   'Manufacturing Director':5,'Healthcare Representative':4,'Manager':3,
                                   'Sales Representative':2,'Research Director':1,'Human Resources':0})
```

```python
hr.MaritalStatus = hr.MaritalStatus.map({'Single':2,'Married':1,'Divorced':0})
```

```python
hr.OverTime = hr.OverTime.map({'Yes':1,'No':0})
```

```python
hr.info()
```

# Feature Selection

```python
hr.drop(['EmployeeCount','EmployeeNumber','Over18','StandardHours'],axis=1, inplace=True)
```

# Model Creation

```python
x = hr.drop('Attrition',axis=1)
y = hr['Attrition']
```

```python
x.head()
```

```python
from collections import Counter
from imblearn.over_sampling import SMOTE
sm = SMOTE()
print('Unbalanced data :', Counter(y))
x_sm,y_sm = sm.fit_resample(x,y)
print('Balanced data :',Counter(y_sm))
```

# Train-Test Split

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_sm,y_sm, random_state=42, test_size=0.25)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
```

```python
dt.fit(x_train,y_train)
```

```python
y_pred= dt.predict(x_test)
y_pred
```

```python
from sklearn.metrics import classification_report, accuracy_score, f1_score
```

```python
print(classification_report(y_test,y_pred))
```

```python
train_pred = dt.predict(x_train)
```

```python
print(classification_report(y_train, train_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

# Random Forest Classifier

```python
rfc = RandomForestClassifier(n_estimators=100)
```

```python
rfc.fit(x_train, y_train)
```

```python
y_prediction = rfc.predict(x_test)
```

```python
print(classification_report(y_test, y_prediction))
```
